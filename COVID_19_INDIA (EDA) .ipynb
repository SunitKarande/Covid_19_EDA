Exploratory Data Analysis
A field of machine learning which understands the data more in depth

EDA is a Data Analytics process to understand the data in depth and to learn the different characteristics of data, often with visual means, This allows tell us to get a fell of our data better and find useful patterns in it.

Why to use EDA ? -Removes irregularities and unnecessary values from data -Helps us prepare our dataset for analysis -Allows a machine learning model to better predict on our dataset -Gives us more accurate results -It also helps us in choosing a better machine learning model

Steps of EDA 1.Data Collection Data collection is an important part of the EDA. It refers to the process of finding and loading data into our system. Good ,Reliable data can be found on various public sites, or can be bought from private organizations. eg:kaggle,github,data.gov,data.gov.in

2.Data Cleaning Data Cleaning refers to the process of removing unwanted variables and values from our dataset and getting rid of any irregularities in it. Such irregularities can disproportionately skew our data and hence adversely effect the results. Removing missing values, outliers and unnecessary rows/columns

#### From the covid dataset delete Sno,Time.ConfirmedIndianNational and FOreign national which are of no use

```python
covid_df.drop(['Sno','Time','ConfirmedIndianNational','ConfirmedForeignNational'], inplace = True,axis = 1)
```

#### There are some irregularities in State/UnionTerritory Column

```python
covid_df['State/UnionTerritory'].unique()
```

```python
## Replacing the *** and cleaning it
covid_df['State/UnionTerritory'] = covid_df['State/UnionTerritory'].str.replace('*','')
```

```python
#covid_df['State/UnionTerritory'].loc(['Cases being reassigned to states'])

covid_df.loc[covid_df['State/UnionTerritory'] == 'Cases being reassigned to states']
```

```python
#covid_df.drop(['Cases being reassigned to states'],axis=0,inplace=True)

covid_df = covid_df[covid_df["State/UnionTerritory"].str.contains("Cases being reassigned to states") == False]
covid_df = covid_df[covid_df["State/UnionTerritory"].str.contains("Unassigned") == False]
covid_df = covid_df[covid_df["State/UnionTerritory"].str.contains("Karanataka") == False]
```

```python
covid_df['State/UnionTerritory'].unique()
```

```python
covid_df.head()
```

#### Change the date column datatype to datetime


```python
covid_df['Date'] = pd.to_datetime(covid_df['Date'], format = '%Y-%m-%d' )
```

```python
covid_df.head()
```

##### Total Active Case
active cases = no of confirmed cases - sum of (cured+deaths)

```python
# Active Cases

covid_df['Active_Cases'] = covid_df['Confirmed']-(covid_df['Cured']+covid_df['Deaths'])
covid_df.tail()
```

```python
# Create pivot table 
statewise = pd.pivot_table(covid_df , values = ['Confirmed','Deaths','Cured'],
                           index = 'State/UnionTerritory', aggfunc = max)
statewise
```

```python
# Recovery Rate 
statewise['Recovery Rate'] = (statewise['Cured'] / statewise['Confirmed']) * 100
```

```python
# Mortality Rate or Death Rate
statewise['Mortality Rate'] = (statewise['Deaths'] / statewise['Confirmed']) * 100
```

```python
statewise = statewise.sort_values(by = 'Confirmed' , ascending = False)
```

```python
statewise.style.background_gradient(cmap = 'YlOrBr')
```

### Top 10 active cases according to states

```python
top_10_active_cases_states = covid_df.groupby(by = 'State/UnionTerritory').max()[['Active_Cases' , 'Date']].sort_values(by = ['Active_Cases'], ascending=False).reset_index()
top_10_active_cases_states
```

```python
fig = plt.figure(figsize=(16,9))
plt.title('Top 10 States with Active Cases' , size = 25)
ax = sns.barplot(data = top_10_active_cases_states.iloc[:10], y = 'Active_Cases', x = 'State/UnionTerritory', linewidth = 2, edgecolor = 'k')
```

```python
bottom_10_active_cases = covid_df.groupby(by = 'State/UnionTerritory').max()[['Active_Cases','Date']].sort_values(by=['Active_Cases'],ascending=True).reset_index()
bottom_10_active_cases
```

```python
fig = plt.figure(figsize=(16,9))
plt.title('Bottom 10 States with Active Cases' , size = 25)
ax = sns.barplot(data = bottom_10_active_cases.iloc[:10], y = 'Active_Cases', x = 'State/UnionTerritory', linewidth = 2, edgecolor = 'k')

plt.xlabel('States' )
plt.xticks(rotation = 90 )
plt.ylabel('Total Active Cases')
plt.show()
```

####   Top states with highest deaths


```python
top_10_states_deaths = covid_df.groupby(by = 'State/UnionTerritory').max()[['Deaths','Date']].sort_values(by = ['Deaths'],ascending=False).reset_index().head(10)
top_10_states_deaths
```

```python

fig = plt.figure(figsize=(18,5))
plt.title('Top 10 States with most Deaths', size = 25)

ax = sns.barplot(data = top_10_states_deaths.iloc[:10],y = 'Deaths', x = 'State/UnionTerritory' , linewidth = 2 , edgecolor = 'k')
plt.xlabel('States')
plt.ylabel('Top Deaths')
plt.show()
```

### States with most vaccinated and least vaccinated


```python

max_vac = vaccine.groupby('State')['Total'].sum().to_frame('Total')
max_vac = max_vac.sort_values('Total', ascending = False)[:5]
max_vac
```

```python
fig = plt.figure(figsize = (10,5))
plt.title('Top 5 Vaccinated States in India' , size = 20)
x = sns.barplot(data = max_vac.iloc[:10] , y=max_vac.Total , x=max_vac.index , linewidth=2 , edgecolor = 'k' )
plt.xlabel('States',fontweight='bold')
plt.ylabel('Vaccination',fontweight='bold')
plt.show()
```

```python
min_vac = vaccine.groupby('State')['Total'].sum().to_frame('Total')
min_vac = min_vac.sort_values('Total', ascending = True)[:5]
min_vac
```

```python
fig = plt.figure(figsize = (15,5))
plt.title('Least 5 Vaccinated States in India' , size = 20)
x = sns.barplot(data = min_vac.iloc[:10] , y=min_vac.Total , x=min_vac.index , linewidth=2 , edgecolor = 'k' )
plt.xlabel('States')
plt.ylabel('Vaccination')
plt.show()
```
